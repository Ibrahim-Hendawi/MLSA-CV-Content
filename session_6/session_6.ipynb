## Image Histogram

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
black_img = cv2.imread('A_black_image.jpg',0)
color_img = cv2.imread('MLSA.png', 0)
plt.imshow(color_img, cmap='gray')
plt.xticks([]), plt.yticks([]) # to hide tick values on x and y axis
plt.show()
```

#### Compute histogram using OpenCV

```python
hist1 = cv2.calcHist([color_img],[0],None,[256],[0,265])
```

#### show the plotting graph of an image

```python
plt.plot(hist1)
plt.show()
```

#### Compute histogram using Numpy

```python
hist2,bins = np.histogram(black_img.ravel(),256,[0,256])
```

```python
plt.plot(hist2)
plt.show()
```

#### using matplotlib.pyplot.hist function

```python
plt.hist(black_img.ravel(),256,[0,256])
plt.show()
```

## Creating histogram for color images

```python
img = cv2.imread('MLSA.png')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

## Thresholding

```python
gray_img = cv2.imread('gray.png', 0)
plt.imshow(gray_img, cmap='gray')
plt.xticks([]), plt.yticks([]) # to hide tick values on x and y axis
plt.show()
```

```python
ret, thresholded = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(thresholded, cmap='gray')
plt.xticks([]), plt.yticks([]) # to hide tick values on x and y axis
plt.show()
```
