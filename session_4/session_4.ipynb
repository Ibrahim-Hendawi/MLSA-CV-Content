### Complete program

```python
img = cv2.imread('1.jpg',0)

cv2.imshow('image',img)

k = cv2.waitKey(0)
if k == 27: # wait for ESC key to exit
    cv2.destroyAllWindows()
elif k == ord('s'): # wait for 's' key to save and exit
    cv2.imwrite('1_save_gray.png',img)
    cv2.destroyAllWindows()
```

### Capture Video from Camera

```python
# Create a VideoCapture object with the device index as an argument
cap = cv2.VideoCapture(0)

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()
    # Our operation on the frame come here
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Display the resulting frame
    cv2.imshow('frame', gray)
    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

### Playing Video from file

```python
# Create a VideoCapture object with the device index as an argument
cap = cv2.VideoCapture('2.avi')

while cap.isOpened():
    # Capture frame-by-frame
    ret, frame = cap.read()

    if not ret:
        break
        
    # Our operation on the frame come here

    # Display the resulting frame
    cv2.imshow('frame', frame)
    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

### Saving a Video

```python
# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'MP4V')
out = cv2.VideoWriter('output.avi',fourcc, 20.0, (640,480))
cap = cv2.VideoCapture('2.avi')
while cap.isOpened():
    # Capture frame-by-frame
    ret, frame = cap.read()

    if not ret:
        break
    
    # Our operation on the frame come here
    frame = cv2.flip(frame, 1)

    # write the fliped frame
    out.write(frame)
    
    # Display the resulting frame
    cv2.imshow('frame', frame)
    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

## Drawing Functions in OpenCV

### Drawing line

```python
# Create a black image
img = np.zeros((512,512,3), np.uint8)

# Draw a diagonal blue line with thickness of 5 px
cv2.line(img,(0,0),(511,511),(255,0,0),100)

cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Drawing Rectangle

```python
# draw rectangle
cv2.rectangle(img,(384,0),(510,128),(0,255,0),3)
```

### Drawing Circle

```python
# draw a circle
cv2.circle(img,(447,63), 63, (0,0,255), -1)
```

### Adding Text

```python
# adding text
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,' MLSA - Club', (10,500), font, 2, (255,255,255), 2)
```

### Display final result

```python
cv2.imshow('image',img)
cv2.imwrite("image.jpg", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
